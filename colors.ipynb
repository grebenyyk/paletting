```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.cm import get_cmap
```

```python
colors1 = sns.color_palette('deep') # almost default matplotlib but better
colors2 = sns.color_palette(['#105775',"#b55020","#c8300e","#4197c8","#69614f"]) # super nice subtle colors
colors3 = sns.color_palette('icefire') # nice and subtle
colors4 = sns.cubehelix_palette(n_colors=10, start=0, rot=0.4, gamma=1.0, hue=0.8, light=0.85, dark=0.15, reverse=True, as_cmap=False)
```

```python
%matplotlib inline
fig, ax = plt.subplots()
fig.set_dpi(150)
ax.set_prop_cycle(color=colors1)
x = np.arange(0,10,0.01)
plt.plot(x, np.sin(x))
plt.plot(x, np.sin(x)-2)
plt.plot(x, np.sin(x)-4)
plt.plot(x, np.sin(x)-6)
plt.plot(x, np.sin(x)-8)
```

```python
%matplotlib inline
sns.palplot(colors1)
plt.title('deep')

sns.palplot(colors2)
plt.title('custom')

sns.palplot(colors3)
plt.title('icefire')

sns.palplot(colors4)
plt.title('some_cubehelix')
```

```python
%matplotlib inline
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(figsize=(6, 5), ncols=2, nrows=2)
fig.set_dpi(150)
N = np.random.random((10,10))

fig1 = ax1.imshow(N, cmap=sns.cubehelix_palette(n_colors=10, start=0, rot=0.4, gamma=1.0, hue=0.8, light=0.85, dark=0.15, reverse=False, as_cmap=True))
fig.colorbar(fig1)

fig2 = ax2.imshow(N, cmap=sns.color_palette('viridis', as_cmap=True))
fig.colorbar(fig2)

fig3 = ax3.imshow(N, cmap=sns.color_palette('mako', as_cmap=True))
fig.colorbar(fig3)

fig3 = ax4.imshow(N, cmap=sns.color_palette('GnBu_d', as_cmap=True))
fig.colorbar(fig3)
```
